```python
1. User Login Check 
Problem Statement: 
Given a username and password, check whether login is successful. 
```

```python
def user_check():
    #take username and password from the user
    input_username = input("Enter username: ")
    input_password = input("Enter password: ")

    #checking the crdentials and then validating
    if input_username == "admin" and input_password == "1234":
        print("Login Successful")
    else:
        print("Invalid Credentials")
#calling the function
user_check()
```

```python
2. Pass / Fail Analyzer 
Problem Statement: 
Analyze student results. 
```

```python
marks = [45, 78, 90, 33, 60]
#initialize an empty count 
pass_count = 0
fail_count = 0

#checking if marks is greater than 50 or not
for mark in marks:
  if mark >= 50:
      pass_count+=1
  else:
      fail_count+=1
#printing the counted values  
print("Total students:",len(marks))
print("Total passed students:",pass_count)
print("Total failed students:",fail_count)
```

```python
3. Simple Data Cleaner 
Problem Statement: 
Clean and standardize user names. 
names = [" Alice ", "bob", " CHARLIE "]
```

```python
names = [" Alice ", "bob", " CHARLIE "]
#intialize an empty list 
cleaned_name=[]
#remove the space and converting chararacters in lower case
for name in names:
    clean=name.strip().lower()
    cleaned_name.append(clean)

print(cleaned_name)
```

```python
4. Message Length Analyzer 
Problem Statement: 
Analyze message sizes. 
messages = ["Hi", "Welcome to the platform", "OK"] 
```

```python
messages = ["Hi", "Welcome to the platform", "OK"] 
#print each message and its length
for message in messages:
    length = len(message)
    print(f"Message: '{message}'| Length: {length}")
#checking the length of message if it's greater than 10 or not
    if length > 10:
        print("This message is longer than 10 characters")
    else:
        print("message lenth is acceptable")
```
