Problem 1: Mobile Recharge Validation System
Problem Statement
Create a function to validate a mobile recharge request.
Rules:
● Recharge amount must be ≥ ₹50
● Valid plans: 199, 299, 399, 599
● User should be allowed to retry if invalid

```python
valid_plans = [199, 299, 399, 599]

# Function to check if the entered amount is a valid recharge
def valid_recharge(amount):

    # Check two conditions:
    # 1. Amount should be at least 50
    # 2. Amount must be one of the predefined valid plans
    if amount >= 50 and amount in valid_plans:
        return True   # Recharge is valid
    else:
        return False  # Recharge is invalid


# Infinite loop to keep asking user until valid amount is entered
while True:

    # Take recharge amount as input from user
    amount = int(input("Enter recharge amount: "))

    # Check if recharge amount is valid
    if valid_recharge(amount):
        print(f"Recharge amount successful of: {amount}")
        break   # Exit loop if valid recharge
    else:
        # If recharge is invalid, show Invalid messages
        print("Invalid amount")
        print("Amount should be greater than 50")

        print(f"Valid plans are: {', '.join(str(p) for p in valid_plans)}")

```

Problem 2: Inventory Reorder Alert System
Problem Statement
Given a dictionary of products and stock quantities:
● Print Reorder Alert if stock < 15
● Print Stock OK otherwise


```python
# Dictionary storing product names as keys and their stock quantities as values
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 8,
    "Monitor": 20,
    "Headphones": 5
}

# Function to check inventory and print reorder alerts
def reorder_alert(inventory):
    # Loop through each product and its quantity in the inventory
    for product, quantity in inventory.items():

        # If stock is less than 15, trigger reorder alert
        if quantity < 15:
            print(f"Reorder alert for {product}: {quantity}")
        else:
            # Otherwise, stock level is sufficient
            print(f"{product}: Stock Ok")

# Call the function to check current inventory
reorder_alert(inventory)
```

Problem 3: Student Result Processing System
Problem Statement
Create a function that:
● Accepts a list of student marks
● Calculates average marks
● Prints Pass if average ≥ 50 else Fail

```python
# Function to calculate average marks and determine result
def student_result(marks):

    # Calculate average of the marks
    average = sum(marks) / len(marks)
    print(f"Average: {average}")

    # Check if student passed or failed
    if average >= 50:
        print("Pass")
    else:
        print("Fail")


# Take input as comma-separated marks from user
marks = input("Enter marks separated by commas: ")

# Convert input string into a list of integers
marks_list = [int(x) for x in marks.split(',')]

# Print the list of marks entered
print(f"Marks are: {marks_list}")

# Call function to evaluate result
student_result(marks_list)

```

Problem 4: Cab Fare Estimator with Retry Option
Problem Statement
Create a cab fare calculator:
● Base fare = ₹50
● ₹12 per km
● Peak hour → 25% extra
● Allow retry using while loop


```python
# Function to calculate cab fare based on distance and peak hour status
def cab_fare(distance, is_peak_hour):

    base_fare = 50  # Fixed starting fare
    fare = base_fare + (distance * 12)  # Add per km charge (₹12 per km)

    # Apply 25% surge pricing during peak hours
    if is_peak_hour:
        fare = fare * 1.25

    return fare


# Loop to allow multiple fare calculations
while True:
    try:
        # Take distance input and convert to float
        distance_input = float(input("Enter travelled distance: "))

        # Ask user if it is peak hour
        peak_hour_str = input("Is it a peak hour? (yes/no): ")

        # Convert input to boolean (True if yes, False otherwise)
        is_peak_hour = (peak_hour_str.lower() == "yes")

        # Calculate fare using the function
        fare = cab_fare(distance_input, is_peak_hour)

        # Display fare formatted to 2 decimal places
        print(f"Total fare: ₹{fare:.2f}")

        # Ask user if they want to calculate another fare
        retry = input("Calculate another fare? (yes/no): ")

        # Exit loop if user does not type "yes"
        if retry.lower() != "yes":
            break

    # Handle invalid numeric input for distance
    except ValueError:
        print("Invalid input. Please enter a valid number for distance.")

    # Handle any unexpected errors
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

Problem 5: Employee Attendance Eligibility
Checker
Problem Statement
Create a function that:
● Accepts attendance list ("P", "A")
● Calculates attendance percentage
● Returns Eligible if ≥ 75% else Not Eligible

```python
attendance = input("Enter attendance (P/A separated by commas): ")

attendance_list = attendance.split(',')

present_days = attendance_list.count('p')
total_days = len(attendance_list)

percentage = (present_days / total_days) * 100

print(f"Present days: {present_days}/{total_days}")
print(f"Attendance: {percentage:.2f}%")

if percentage >= 75:
    print("Status: Eligible")
else:
    print("Status: Not Eligible")
```

Problem 6: Password Strength Checker
Problem Statement
Create a function to check password strength.
Rules:
● Minimum length = 8
● Must contain at least one digit
● Must contain at least one special character (@#$)


```python
def password_checker(password):
  issue=[]
  if len(password) < 8:
    issue.append("Password is too short (min 8 characters)")

  has_digit = False
  for char in password:
    if char.isdigit():
      has_digit = True
      break
  if not has_digit:
    issue.append("Password requires at least one digit")

  has_special_char = False
  special_chars = "@#$%^&[]{}|/"
  for char in password:
    if char in special_chars:
      has_special_char = True
      break
  if not has_special_char:
    issue.append("Password requires at least one special character (e.g., @#$%) ")

  if not issue:
    return "strong"
  else:
    print("Password issues:", ", ".join(issue))
    return "weak"


while True:
  password=input("enter password: ")
  result=password_checker(password)
  print(result)
  retry = input("Check another password? (yes/no): ")
  if retry.lower() != "yes":
    break
    #too tired of writing comments , preferred using chatgpt for comments writing
```
