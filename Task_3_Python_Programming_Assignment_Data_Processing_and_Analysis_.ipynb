Problem Statement 1: Employee Performance Bonus Eligibility
Description:
A company evaluates employee performance scores at the end of the year. You are given a
dictionary containing employee names and their performance scores.


```python
# Employee names mapped to their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest score
max_score = max(employees.values())

# Store employees who have the highest score
top_performers = []

# Identify top-performing employees
for name, score in employees.items():
    if score == max_score:
        top_performers.append(name)

# Display the result
print("TOP Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", str(max_score) + ")")
```

Problem Statement 2: Search Query Keyword Analysis
Description:
An e-commerce website stores customer search queries. You are given a search query
sentence entered by a user.


```python
# Convert sentence to lowercase and split into words
sentence = "Buy mobile phone buy phone online"
sentence = sentence.lower().split()

# Dictionary to store word frequencies
word_count = {}

print(sentence)

# Count occurrences of each word
for word in sentence:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Store words that appear more than once
result = {}
for word, count in word_count.items():
    if count > 1:
        result[word] = count

# Display repeated words with their counts
print(result)
```

Problem Statement 3: Sensor Data Validation
Description:
A factory collects sensor readings every hour. Each reading is stored in a list where the index
represents the hour and the value represents the sensor reading.

```python
# List of sensor readings recorded hourly
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Store readings with even values
valid_readings = []

# Collect index and value for even readings
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))

# Display results
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem Statement 4: Email Domain Usage Analysis
Description:
A company wants to analyze which email providers its users are using. You are given a list of
employee email IDs.


```python

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to store domain frequency
domain_counts = {}

# Extract and count email domains
for email in emails:
    parts = email.split('@')
    if len(parts) == 2:
        domains = parts[1]
        if domains in domain_counts:
            domain_counts[domains] += 1
        else:
            domain_counts[domains] = 1

# Calculate total number of emails
total_employees_mails = len(emails)

# Compute and display percentage for each domain
for domains, count in domain_counts.items():
    percentage = (count / total_employees_mails) * 100
    print(f"{domains}:{percentage: 0f}%")
```

Problem Statement 5: Sales Spike Detection
Description:
A retail company tracks daily sales. Sudden spikes in sales may indicate promotions or unusual
activity.

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average daily sales
average_daily = sum(sales) / len(sales)

# Set threshold at 30% above average
threshold = average_daily * 1.3

# Identify days where sales exceed the threshold
for i in range(len(sales)):
    sale = sales[i]
    if sale > threshold:
        print(f"Day {i+1}: {sale}")


```

Problem Statement 6: Duplicate User ID Detection
Description:
A system stores user IDs during registration. Duplicate IDs can cause data integrity issues.


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to store frequency of each user ID
id_counts = {}

# Count occurrences of each user ID
for user_id in user_ids:
    if user_id in id_counts:
        id_counts[user_id] += 1
    else:
        id_counts[user_id] = 1

# Display user IDs that appear more than once
for user_id, count in id_counts.items():
    if count > 1:
        print(f"{user_id} → {count} times")
```
