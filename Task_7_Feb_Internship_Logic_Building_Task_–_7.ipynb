Problem 1: Smart Light Controller
Problem Statement
Create a class to control a smart light that can be turned ON or OFF and display its
status.


```python
class smartlight:
  # Initialize the smartlight with a given name
  def __init__(self, name):
    self.name=name # Store the name of the light
    self.status="OFF" # Set initial status to OFF

  # Method to turn the light ON
  def turn_on(self):
    self.status="ON" # Change status to ON

  # Method to turn the light OFF
  def turn_off(self):
    self.status="OFF" # Change status to OFF

  # Method to display the current status of the light
  def display_status(self):
    print(f"{self.name} is {self.status}") # Print the light's name and status


# Create an instance of the smartlight class
light = smartlight("Bedroom Light")

# Turn the bedroom light on
light.turn_on()
# Display the current status of the bedroom light
light.display_status()
```

Problem 2: Employee ID Card System
Problem Statement
Create a class that stores employee name, ID, and department and displays ID card
details.

```python
class employee:
  # Initialize the employee with name, ID, and department
  def __init__(self,name,id,department):
    self.name=name # Store the employee's name
    self.id=id # Store the employee's ID
    self.department=department # Store the employee's department

  # Method to display the employee's ID card details
  def display_id_card(self):
     print("Employee ID Card") # Print a header for the ID card
     print(f"Employee Name: {self.name}") # Print the employee's name
     print(f"Employee ID: {self.id}") # Print the employee's ID
     print(f"Department: {self.department}") # Print the employee's department


# Create an instance of the employee class
emp=employee("Rahul","EMP102","IT")
# Display the ID card for the created employee
emp.display_id_card()
```

Problem 3: Mobile Contact Record
Problem Statement
Create a class to store contact name and phone number and display contact
information

```python
class mobile_record:
  # Initialize the mobile record with name and phone number
  def __init__(self,name,phone_number):
    # Convert phone_number to string before checking length
    if not str(phone_number).isdigit() or len(str(phone_number))!=10:
      raise ValueError("Invalid Number")

    self.name=name # Store the contact name
    self.number=phone_number # Store the phone number

  # Method to display the contact information
  def display_contact(self):
    print("Contact Saved") # Print a header
    print(f"Name: {self.name}") # Print the contact name
    print(f"Phone:{self.number}") # Print the phone number


# Create an instance of the mobile_record class
info=mobile_record("Anita",9876543210)
# Display the contact information
info.display_contact()
```

Problem 4: Product Price Tag Generator
Problem Statement
Create a class that stores product name and price and prints a formatted price tag.
Real-Time Use
Retail billing systems


```python
class Product:

    def __init__(self, name, price):
        if price < 0:
            raise ValueError("Price cannot be negative")

        self.name = name # Store the product name
        self.price = price # Store the product price

    def display_price_tag(self):
        print(f"Product: {self.name}") # Print the product name
        print(f"Price: ₹{self.price:.2f}") # Print the formatted price

p1 = Product("Headphones", 2499) # Create an instance of the Product class
p1.display_price_tag() # Display the price tag
```

Problem 5: Movie Rating Display System
Problem Statement
Create a class that stores movie name and rating and displays movie details.


```python
class Movie:

    def __init__(self, name, rating):
        if rating < 0 or rating > 5:
            raise ValueError("Rating must be between 0 and 5")

        self.name = name # Store the movie name
        self.rating = rating # Store the movie rating

    def display_movie(self):
        print(f"Movie: {self.name}") # Print the movie name
        print(f"Rating: {self.rating} / 5") # Print the movie rating

m1 = Movie("Inception", 4.8) # Create an instance of the Movie class
m1.display_movie() # Display the movie details
```

Problem 6: Delivery Address Manager
Problem Statement
Create a class that stores customer name and delivery address and prints delivery
details.


```python
class Delivery:

    def __init__(self, customer_name, address):
        if not customer_name:
            raise ValueError("Customer name cannot be empty")

        if not address:
            raise ValueError("Address cannot be empty")

        self.customer_name = customer_name # Store the customer's name
        self.address = address # Store the delivery address

    def display_delivery_details(self):
        print("Delivery Details") # Print a header for delivery details
        print(f"Customer: {self.customer_name}") # Print the customer's name
        print(f"Address: {self.address}") # Print the delivery address

d1 = Delivery("Suman", "Hyderabad") # Create an instance of the Delivery class
d1.display_delivery_details() # Display the delivery details
```
