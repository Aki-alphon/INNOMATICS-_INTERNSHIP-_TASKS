Problem 1: Social Media – Post Engagement Analyzer
Problem Statement
Create a function to analyze engagement on a social media post.
Rules
● Calculate total likes
● If Total Likes ≥ 1000 → Viral Post
● Otherwise → Normal Engagement

```python
likes = [200, 350, 400, 100]  # Likes from different sources for the social media post

def analyze_like(likes):
  total_likes = sum(likes)  # Calculate the sum of all likes
  if total_likes >= 1000:  # Check if total likes are 1000 or more
    status = "Viral Post"  # Assign 'Viral Post' status
  else:
    status = "Normal Engagement"  # Assign 'Normal Engagement' status
  return total_likes, status  # Return the total likes and status

total_likes,status=analyze_like(likes)  # Call the function with the likes list

print(f"Total Likes:{total_likes}")  # Print the total number of likes
print(f"Post Status:{status}")  # Print the engagement status of the post
```

Problem 2: Healthcare – Medicine Stock Alert System
Problem Statement
Create a function to monitor medicine stock levels.
Rules
● Stock < 10 → Low Stock Alert
● Stock ≥ 10 → Stock Sufficient

```python
def medicine_stock(stocks):

  if stocks < 10 and stocks > 0: # Check if stock is low but not zero or negative
    status_med = "Low Stock Alert" # Assign 'Low Stock Alert' status
  elif stocks >= 10: # Check if stock is sufficient (10 or more)
    status_med = "Stock Sufficient" # Assign 'Stock Sufficient' status
  elif stocks == 0: # Check if stock is zero
    status_med = "Out of Stock" # Assign 'Out of Stock' status
  elif stocks < 0: # Check if stock is negative (invalid)
    status_med = "Invalid Stock" # Assign 'Invalid Stock' status
  else: # Catch any other unexpected cases
    status_med = "Unknown Status" # Assign 'Unknown Status' status

  return status_med,stocks # Return the status and the stock value



stocks = [-5, 0, 1, 5, 9, 10, 15, 100] # Define a list of stock levels for testing

# Iterate through the list of stocks and print the status for each
for stock_level in stocks:
  status, stock_value = medicine_stock(stock_level) # Call the function for each stock level
  print(f"Stock: {stock_value}, Status: {status}") # Print the stock level and its corresponding status
```

Problem 3: Agriculture – Rainfall Adequacy Checker
Problem Statement
Create a function that analyzes seasonal rainfall data.
Rules
● Calculate average rainfall
● Average ≥ required level → Adequate Rainfall
● Otherwise → Inadequate Rainfall

```python
def rainfall_checker(rainfall,required_level):
  average_rainfall = sum(rainfall)/len(rainfall) # Calculate the average rainfall from the list
  if average_rainfall >= required_level: # Check if the average rainfall meets the required level
    status_rainfall = "Adequate Rainfall" # Assign 'Adequate Rainfall' status
  else:
    status_rainfall = "Inadequate Rainfall" # Assign 'Inadequate Rainfall' status
  return average_rainfall , status_rainfall # Return the calculated average and the status

rainfall=[10,20,30,40,50,60] # Define a list of rainfall data
required_level=float(input("Enter Required Level: ")) # Get the required rainfall level from the user
average_rainfall,status_rainfall=rainfall_checker(rainfall,required_level) # Call the function to check rainfall adequacy
print(f"Average Rainfall:{average_rainfall}") # Print the average rainfall
print(f"Rainfall Status:{status_rainfall}") # Print the rainfall status
```

Problem 6: Agriculture – Premium Crop Price Filter
Problem Statement
Create a function that filters crop prices above a premium threshold.
Rules
● Price > ₹2000 → Premium Crop
● Else → Regular Crop


```python
def filter_crop(prices,threshold):
  premium_list=[] # Initialize an empty list to store premium crop prices
  regular_list=[] # Initialize an empty list to store regular crop prices

  for price in prices:
   if price < 0:
    print(f"Warning: Invalid price detected -> {price}") # Warn about invalid prices
    continue # Skip to the next price if invalid

   if price >= threshold:
    premium_list.append(price) # Add price to premium list if it meets or exceeds the threshold
   else:
    regular_list.append(price) # Add price to regular list if it's below the threshold

  return premium_list,regular_list # Return both lists of filtered prices

crop_prices = [1500, 2500, -300, 3200, 1800] # Define a list of crop prices
threshold = 2000 # Define the premium price threshold
premium,regular=filter_crop(crop_prices,threshold) # Call the function to filter crop prices
print(f"Premium Crop: {premium}") # Print the list of premium crop prices
```

Problem 7: System Monitoring – Application Health
Checker
Problem Statement
Create a function to evaluate application health based on error count.
Rules
● Errors = 0 → Healthy
● Errors ≤ 5 → Minor Issues
● Errors > 5 → Critical Issues


```python
def health_check(error_count): # Define the health_check function that takes error_count as input

 if error_count < 0: # Check if error count is negative (invalid)
    return "Invalid Error Count" # Return 'Invalid Error Count' status

 elif error_count == 0: # Check if error count is zero
    return "Healthy" # Return 'Healthy' status

 elif error_count <= 5: # Check if error count is between 1 and 5 (inclusive)
    return "Minor Issues" # Return 'Minor Issues' status

 else: # If error count is greater than 5
    return "Critical Issues" # Return 'Critical Issues' status

try:
   error_count = int(input("Enter Error Count:")) # Prompt user for error count and convert to integer
   # Assign the returned value of health_check to a variable named 'status'
   status = health_check(error_count) # Call health_check function with user input
   print(f"Error Count: {error_count}") # Print the entered error count
   print(f"System Status: {status}") # Print the system status based on error count
except ValueError:
    print("Invalid Input. Please enter a valid integer.") # Handle cases where input is not an integer
```

Problem 8: Banking – Daily Transaction Limit Checker
Problem Statement
Create a function to validate daily transaction amount.
Rules
● Daily limit = ₹50,000
● Amount within limit → Approved
● Amount exceeds limit → Rejected


```python
def amount_checker(amt,daily_limit):
  if amt < 0:
    return "Invalid amount" # Return 'Invalid amount' for negative input

  if amt <= daily_limit:
    return "Approved" # Return 'Approved' if amount is within daily limit
  else:
    return "Rejected" # Return 'Rejected' if amount exceeds daily limit

try:
   amt = float(input("Enter Transaction Amount: ")) # Prompt user for transaction amount and convert to float
   daily_limit = 50000 # Define the daily transaction limit
   print(f"Transaction Amount: {amt}") # Print the entered transaction amount
   print(f"Transaction Status: {amount_checker(amt,daily_limit)}") # Call the function and print the transaction status
except:
  print("Invalid") # Handle non-numeric input
```

Problem 9: E-Learning – Student Attendance Eligibility
System
Problem Statement
Create a function to calculate attendance percentage and check exam eligibility.
Rules
● Attendance ≥ 75% → Eligible
● Attendance < 75% → Not Eligible


```python
def attendance_checker(attendance):
    if not attendance:
        return 0, "Invalid Attendance" # Return 0 and 'Invalid Attendance' if attendance list is empty

    present_count = 0 # Initialize a counter for present days

    for day in attendance:
        day = day.strip().upper() # Clean and standardize the attendance entry

        if day not in ["P", "A"]:
            return 0, "Invalid Attendance" # Return 0 and 'Invalid Attendance' for invalid entries

        if day == "P":
            present_count += 1 # Increment present count if the day is 'P'

    percentage = (present_count / len(attendance)) * 100 # Calculate attendance percentage

    if percentage >= 75:
        status = "Eligible" # Set status to 'Eligible' if percentage is 75% or more
    else:
        status = "Not Eligible" # Set status to 'Not Eligible' otherwise

    return percentage, status # Return the calculated percentage and status


# Test
attendance_list = ["P","A","P","A","P","A","P","A"] # Example attendance list

percentage, status = attendance_checker(attendance_list) # Call the function to get percentage and status

print(f"Attendance Percentage: {percentage:.2f}%") # Print the formatted attendance percentage
print(f"Exam Eligibility: {status}") # Print the exam eligibility status
```

Problem 10: Smart Electricity Bill Analyzer
Problem Statement
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units
consumed.
Billing Rules
● First 100 units → ₹3 per unit
● Next 100 units (101–200) → ₹5 per unit
● Above 200 units → ₹7 per unit
Usage Classification
● Total bill < ₹500 → Low Usage
● ₹500 – ₹1500 → Moderate Usage
● ₹1500 → High Usage


```python
def bill_analyzer(units):
    if units < 0:
        return 0, "Invalid Usage" # Return 0 and 'Invalid Usage' for negative units

    if units <= 100:
        bill = units * 3 # Calculate bill for first 100 units
        usage = "Low Usage" # Classify as 'Low Usage'

    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5) # Calculate bill for units between 101 and 200
        usage = "Moderate Usage" # Classify as 'Moderate Usage'

    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7) # Calculate bill for units above 200
        usage = "High Usage" # Classify as 'High Usage'

    return bill, usage # Return the calculated bill and usage category


try:
    units = int(input("Enter Units: ")) # Prompt user for unit input and convert to integer
    billing, use = bill_analyzer(units) # Call bill_analyzer function
    print(f"Price to Pay: {billing}") # Print the calculated bill
    print(f"Usage Category: {use}") # Print the usage category
except ValueError:
    print("Invalid Input. Please enter a valid integer.") # Handle non-integer input
```
