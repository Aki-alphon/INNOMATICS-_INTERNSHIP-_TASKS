Problem 1: Unique Words in a Sentence
Problem Statement
Given a sentence, identify all unique words and display:
1. The count of unique words
2. The unique words themselves

```python
input_sentence = input("Enter sentence: ")
#spliting the sentence into words
words = input_sentence.split()
#Convert list to set to remove duplicates
unique_word = set(words)
#count of unique words
word_length = len(unique_word)
#printing the count and words themselves
print("Count of Unique words:",word_length)
print("unique words are:",unique_word)
```

Problem 2: Highest Salary from Employee Data
Problem Statement
Store employee names and their salaries, and determine which employee has the highest
salary.


```python
employees = {
"Ravi": 75000,
"Anita": 68000,
"Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)

print(f"Highest Salary: {highest_employee} - {employees[highest_employee]}")
```

Problem 4: Count Products Above a Price Threshold
Problem Statement
Count how many products have a price greater than 1000.

```python
prices = [450, 1200, 899, 1500, 300]
#initializing the count to 0
count=0
# comparing each value in list using for loop
for price in prices:
  if price > 1000:
    #incring the count value if price is greater than 1000 found
   count += 1
#printing the count value
print("Products above 1000:",count)
```

Problem 5: Calculate Attendance Percentage
Problem Statement
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

Problem 7: Count Character Frequency
Problem Statement
Count how many times each character appears in a given string.

```python
get_string=input("Enter the sentence: ")
#intinalizing empty dictinary
freq={}
#for loop counts key characters in the get_string
for char in get_string:
  #if characters get repeated then its value in freq is increased by 1
  if char in freq:
    freq[char]+=1
    #otherwise no repeatation of key then its value is 1
  else:
    freq[char]=1
    #printing the dictinary
print("Frequency:",freq)
```

Problem 10: Calculate Average Marks
Problem Statement
Calculate the average marks of a student from a list of marks.

```python
marks_input = input("Enter marks separated by commas (e.g., 40,50): ")
#taken input is stored in list
marks = [int(mark.strip()) for mark in marks_input.split(',')]
#calculating the total
total = sum(marks)
#finding average using formula
average = total / len(marks)
#printing the average
print("Average Marks:", average)
```
